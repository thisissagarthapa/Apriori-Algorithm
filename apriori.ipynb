```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
market=[]
for i in range(0,datasets.shape[0]):
    cus=[]
    for j in datasets.columns:
        if type(datasets[j][i])==str:
            cus.append(datasets[j][i])
    market.append(cus)
            
market
```

```python
import collections
```

```python
# collections.Counter(["a","b","c","a"])
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
        
    
```

```python
l
```

```python
p=collections.Counter(l)
```

```python
p.keys()
```

```python
p.values()
```

```python
dict={
    "items":p.keys(),
    "values":p.values()
}
```

```python
df=pd.DataFrame(dict)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr=TransactionEncoder()
```

```python
df1=pd.DataFrame(tr.fit_transform(market),columns=tr.columns_)
df1
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
ap=apriori(df1,min_support=0.05,use_colnames=True,max_len=3).sort_values(by=["support"])
ap
```
